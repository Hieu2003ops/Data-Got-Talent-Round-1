```python
import re
import unicodedata

def remove_accents(input_str):
    nfkd_form = unicodedata.normalize('NFKD', input_str)
    return "".join([c for c in nfkd_form if not unicodedata.combining(c)])

customer['address'] = customer['address'].astype(str)
customer['address'] = customer['address'].apply(remove_accents)
```

### Thay format ch√≠nh ƒë·ªÉ x·ª≠ l√Ω address

```python
# ƒê·ªãnh nghƒ©a h√†m ƒë·ªÉ x·ª≠ l√Ω c·ªôt 'address' theo y√™u c·∫ßu:
# - N·∫øu c√≥ nhi·ªÅu h∆°n 2 d·∫•u ph·∫©y, x√≥a d·∫•u ph·∫©y ·ªü gi·ªØa.
# - N·∫øu kh√¥ng c√≥ d·∫•u ph·∫©y ·ªü v·ªã tr√≠ cu·ªëi c√πng, th√™m d·∫•u ph·∫©y v√† "ƒë√† n·∫µng".
def process_address(address):
    # ƒê·∫øm s·ªë d·∫•u ph·∫©y
    comma_count = address.count(',')
    
    # X·ª≠ l√Ω tr∆∞·ªùng h·ª£p c√≥ nhi·ªÅu h∆°n 2 d·∫•u ph·∫©y
    if comma_count > 2:
        # T√°ch chu·ªói th√†nh c√°c ph·∫ßn v√† lo·∫°i b·ªè ph·∫ßn t·ª≠ th·ª© hai (d·∫•u ph·∫©y th·ª© hai)
        parts = address.split(',')
        address = ','.join(parts[:2] + parts[3:])
    # X·ª≠ l√Ω tr∆∞·ªùng h·ª£p kh√¥ng c√≥ d·∫•u ph·∫©y ·ªü cu·ªëi
    elif comma_count < 2:
        # Th√™m ",ƒë√† n·∫µng" v√†o cu·ªëi chu·ªói
        address += ',ƒë√† n·∫µng'
        
    return address

# √Åp d·ª•ng h√†m ƒë√£ ƒë·ªãnh nghƒ©a l√™n c·ªôt 'address' c·ªßa DataFrame customer
customer['address'] = customer['address'].apply(process_address)
```

```python
def remove_danang_abbreviations(address):
    # ƒê·ªãnh nghƒ©a danh s√°ch c√°c t·ª´ vi·∫øt t·∫Øt c·∫ßn lo·∫°i b·ªè
    danang_abbreviations = ['ƒën', 'tp ƒë√† n·∫µng', 'da nang','tp. da nang',"ƒëa nang","tp ƒëa nang","tp.","tp"]
    
    # Ki·ªÉm tra v√† thay th·∫ø c√°c t·ª´ vi·∫øt t·∫Øt b·∫±ng chu·ªói r·ªóng
    for abbr in danang_abbreviations:
        address = address.replace(abbr, '')
    
    # Lo·∫°i b·ªè c√°c d·∫•u ph·∫©y th·ª´a v√† kho·∫£ng tr·∫Øng
    address_clean = ','.join([part.strip() for part in address.split(',') if part.strip()])
    
    return address_clean

# √Åp d·ª•ng h√†m ƒë√£ ƒë·ªãnh nghƒ©a l√™n c·ªôt 'address' c·ªßa DataFrame customer
customer['address'] = customer['address'].apply(remove_danang_abbreviations)
```

```python
import re

def remove_accents(input_str):
    # B·∫£ng ƒë·ªïi ch·ªØ c√≥ d·∫•u sang kh√¥ng d·∫•u
    replacements = (
        ("√†|√°|·∫£|√£|·∫°|ƒÉ|·∫±|·∫Ø|·∫≥|·∫µ|·∫∑|√¢|·∫ß|·∫•|·∫©|·∫´|·∫≠", "a"),
        ("√®|√©|·∫ª|·∫Ω|·∫π|√™|·ªÅ|·∫ø|·ªÉ|·ªÖ|·ªá", "e"),
        ("√¨|√≠|·ªâ|ƒ©|·ªã", "i"),
        ("√≤|√≥|·ªè|√µ|·ªç|√¥|·ªì|·ªë|·ªï|·ªó|·ªô|∆°|·ªù|·ªõ|·ªü|·ª°|·ª£", "o"),
        ("√π|√∫|·ªß|≈©|·ª•|∆∞|·ª´|·ª©|·ª≠|·ªØ|·ª±", "u"),
        ("·ª≥|√Ω|·ª∑|·ªπ|·ªµ", "y"),
    )
    
    for pattern, repl in replacements:
        input_str = re.sub(pattern, repl, input_str)
        
    return input_str.lower()

tinh_thanh_list = ["quang tri", "quang ngai", "hue",  
                   "quang binh", "binh dinh", "dong nai",  
                   "khanh hoa", "hai duong", "nghe an",
                   "gia lai", "kon tum", "dak lak", "ha tinh",
                   "ha noi", "tp ho chi minh", "quang nam","ho chi minh","da nang"]
                   
quan_da_nang = ["son tra", "ngu hanh son", "cam le", 
                "thanh khe", "lien chieu", "hoa vang","hai chau"]
                

def split_address(address, quan_list, tinh_thanh_list):
    address_norm = remove_accents(address)

    # Lo·∫°i b·ªè t·ª´ "qu·∫≠n" n·∫øu c√≥ trong t√™n ƒë∆∞·ªùng
    address_norm = re.sub(r'\bquan\b', '', address_norm).strip()

    # X√°c ƒë·ªãnh th√†nh ph·ªë
    thanhpho = None
    for tinh_thanh in tinh_thanh_list:
        tinh_thanh_norm = remove_accents(tinh_thanh)
        if tinh_thanh_norm in address_norm:
            thanhpho = tinh_thanh
            address_norm = address_norm.replace(tinh_thanh_norm, "").strip()
            break
       # N·∫øu kh√¥ng t√¨m th·∫•y trong danh s√°ch t·ªânh, m·∫∑c ƒë·ªãnh l√† ƒê√† N·∫µng
    if not thanhpho:
        thanhpho = "ƒë√† n·∫µng"

    # X√°c ƒë·ªãnh qu·∫≠n v√† ƒë∆∞·ªùng
    quan = None
    duong = None

    # N·∫øu ƒë·ªãa ch·ªâ kh√¥ng thu·ªôc ƒê√† N·∫µng, ƒë·∫∑t ƒê∆∞·ªùng l√† "Ngo·∫°i t·ªânh"
    if thanhpho != "ƒë√† n·∫µng":
        quan = "ngo·∫°i t·ªânh"
        duong = "ngo·∫°i t·ªânh"
    else:
        # T√¨m v√† lo·∫°i b·ªè t√™n qu·∫≠n t·ª´ ƒë·ªãa ch·ªâ chu·∫©n h√≥a
        for q in quan_list:
            q_norm = remove_accents(q)
            if q_norm in address_norm:
                quan = q
                address_norm = address_norm.replace(q_norm, "").strip()
                break
         # Lo·∫°i b·ªè t·ª´ "ƒë√† n·∫µng" n·∫øu c√≥ trong chu·ªói ƒë·ªãa ch·ªâ chu·∫©n h√≥a
        da_nang_norm = remove_accents(",ƒë√† n·∫µng")
        address_norm = address_norm.replace(da_nang_norm, "").strip()

        # ƒê∆∞·ªùng ch·ªâ l√† ph·∫ßn c√≤n l·∫°i c·ªßa ƒë·ªãa ch·ªâ sau khi lo·∫°i b·ªè qu·∫≠n v√† th√†nh ph·ªë
        # duong = address_norm
        duong = re.sub(r'^[,\s]*|[,\s]*$', '', address_norm)
    return duong, quan, thanhpho

# √Åp d·ª•ng h√†m
customer['ƒê∆∞·ªùng'], customer['Qu·∫≠n'], customer['Th√†nh Ph·ªë'] = zip(*customer['address'].apply(lambda x: split_address(x, quan_da_nang, tinh_thanh_list)))

# Xem k·∫øt qu·∫£ sau khi t√°ch
customer[['address', 'ƒê∆∞·ªùng', 'Qu·∫≠n', 'Th√†nh Ph·ªë']]
```

### ƒê·∫∑t l·∫°i c√°c orderid l·∫ßn l∆∞·ª£t l√† Web01,  Web02 ,... cho c√°c orderid null

```python
# L·∫•y ch·ªâ s·ªë cho c√°c b·∫£n ghi m√† 'orderid' l√† null v√† s·∫Øp x·∫øp theo 'customerid' v√† 'date'
null_orderid_records = tickets[tickets['orderid'].isnull()].sort_values(by=['customerid', 'date'])

# T·∫°o m·ªôt bi·∫øn ƒë·ªÉ theo d√µi orderid hi·ªán t·∫°i v√† nh√≥m hi·ªán t·∫°i
current_orderid = 0
current_group = None

# T·∫°o orderid m·ªõi cho m·ªói b·∫£n ghi c√≥ orderid null
for index, row in null_orderid_records.iterrows():
    # Ki·ªÉm tra xem nh√≥m hi·ªán t·∫°i c√≥ kh√°c v·ªõi nh√≥m tr∆∞·ªõc ƒë√≥ kh√¥ng (d·ª±a tr√™n customerid v√† date)
    group = (row['customerid'], row['date'])
    if group != current_group:
        current_group = group
        current_orderid += 1  # Ch·ªâ tƒÉng orderid n·∫øu ch√∫ng ta c√≥ m·ªôt nh√≥m m·ªõi

    # ƒê·ªãnh d·∫°ng orderid m·ªõi v√† c·∫≠p nh·∫≠t trong DataFrame g·ªëc
    new_orderid = 'WEB' + str(current_orderid).zfill(2)
    tickets.at[index, 'orderid'] = new_orderid

# Lo·∫°i b·ªè c·ªôt group_id n·∫øu n√≥ t·ªìn t·∫°i
if 'group_id' in tickets.columns:
    tickets.drop('group_id', axis=1, inplace=True)


```
